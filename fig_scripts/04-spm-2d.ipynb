```python
def color2gray(x):
    if len(x) == 3:
        r, g, b = x
        a = 1
    elif len(x) == 4:
        r, g, b, a = x
    else:
        raise ValueError("Incorrect tuple length")
    return (r * 0.299 + 0.587*g + 0.114*b) * a

```
