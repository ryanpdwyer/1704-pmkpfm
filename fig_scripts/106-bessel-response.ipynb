```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
from __future__ import division, print_function
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import h5py
# import nufftpy
from scipy import special


def jinc(x):
    return np.where(x < 1e-12, 1, 2 * special.j1(x) / x)
```

```python
nu = np.linspace(0, 0.25)
```

```python
plt.rc('font', size=9)
fig, ax = plt.subplots(figsize=(2.5, 2.5), subplot_kw=dict(facecolor='0.9'))
fig.patch.set_facecolor('none')

ax.grid(which='both', color='1', linestyle='-', linewidth=0.75)


ax.axhline(0.95, color='k', linestyle=':')
l1, = ax.plot(nu, special.j0(2*np.pi*nu))
l2, = ax.plot(nu, jinc(2*np.pi*nu))
ax.set_xticks(np.arange(0.05, 0.26, 0.1), minor=True)
ax.text(0.23, 0.56, "$H_0$", color=l1.get_color())
ax.text(0.23, 0.77, "$H_1$", color=l2.get_color())
ax.set_xlabel(u"Normalized frequency $\\nu_y$ [$A_\\mathrm{pm}^{-1}$]")
ax.set_ylabel(u"Response")
```
